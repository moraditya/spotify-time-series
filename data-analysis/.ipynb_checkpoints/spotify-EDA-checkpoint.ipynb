#### Since the genres were fetched from artists names, which can be more than 1 for one song, the code returned, on many instances, more than 1 genre for 1 song. Therefore, before encoding this categorical column, aggregating genres into broader categories and then choosing the mode of the many genres for that 1 row is appropriate. If modes are equal, use the 1st genre.

```python
def convert_to_list(genre_str):
    # try to convert the genre string to a list using ast.literal_eval
    try:
        return ast.literal_eval(genre_str)
    except (ValueError, SyntaxError):
        # if conversion fails, return the genre string as a single-item list
        return [genre_str]

# apply the conversion function to the 'genres' column of the dataframe
streams_df['genres'] = streams_df['genres'].apply(convert_to_list)
```

```python
# print the unique genres, sorted
def flatten(nested_list):
    """Flatten a nested list."""
    for item in nested_list:
        if isinstance(item, Iterable) and not isinstance(item, str):
            yield from flatten(item)
        else:
            yield item

unique_genres = set()

for genres_list in streams_df['genres']:
    # flatten the entire structure and add each genre to the set
    for genre in flatten(genres_list):
        unique_genres.add(genre)

for genre in sorted(unique_genres):
    print(genre)
```

```python
# extract country names out of genre names

def remove_country_from_genre(genre):
    # define a list of country names to remove
    locations = [
        'swedish', 'norwegian', 'german', 'uk', 'british', 'french', 'australian', 'canadian',
        'american', 'scottish', 'japanese', 'italian', 'spanish', 'mexican', 'brazilian', 
        'irish', 'russian', 'korean', 'dutch', 'greek', 'icelandic', 'finnish', 'argentine', 
        'argentinian', 'nashville', 'austrian', 'hong kong', 'indian', 'indonesian', 'danish', 
        'belgian', 'chilean', 'portuguese', 'turkish', 'arabic', 'venezuelan', 'malaysian', 
        'african', 'alabama', 'california', 'atlanta', 'berlin', 'boston', 'brooklyn', 
        'louisiana', 'detroit', 'seattle', 'liverpool', 'london', 'hollywood','gujarati',
        'texas', 'new york', 'new orleans', 'philly', 'vintage', 'melbourne'
    ]
    
    # use regex to remove any matching location names
    for loc in locations:
        genre = re.sub(rf'\b{loc}\b', '', genre)
    
    # clean up any extra whitespace and return the updated genre
    return genre.strip()

# apply the function to clean up the genres in the streams_df_broader dataframe
streams_df_broader['cleaned_broad_genre'] = streams_df_broader['broad_genre'].apply(
    lambda genre_list: [remove_country_from_genre(genre) for genre in genre_list]
)

# flatten the list if there are multiple genres in a single entry, and remove empty strings
streams_df_broader['cleaned_broad_genre'] = streams_df_broader['cleaned_broad_genre'].apply(
    lambda genre_list: [genre for genre in genre_list if genre]  # Filter out any empty strings
)
```

```python
def flatten(nested_list):
    """Flatten a nested list."""
    for item in nested_list:
        if isinstance(item, Iterable) and not isinstance(item, str):
            yield from flatten(item)
        else:
            yield item

unique_genres = set()

for genres_list in streams_df_broader['cleaned_broad_genre']:
    # flatten the entire structure and add each genre to the set
    for genre in flatten(genres_list):
        unique_genres.add(genre)

for genre in sorted(unique_genres):
    print(genre)
```

```python
# drop genres and broad_genres from streams_df_broader dataframe

streams_df_broader = streams_df_broader.drop(columns = ['genres','broad_genre'])
streams_df_broader.head()
```

```python
# function to select the most frequent genre, and if there's a tie, pick the first genre in the list

def pick_most_frequent_genre(genre_list):
    
    # if there's only 1 genre, pick that one
    if len(genre_list) == 1:
        return genre_list[0]
    
    # if there are more than 1 genres, find the mode
    elif len(genre_list) > 1:
        genre_counts = Counter(genre_list)
        max_count = max(genre_counts.values())
        most_frequent_genres = [genre for genre, count in genre_counts.items() if count == max_count]
        
        # if there's a tie with the modes, pick the first genre
        for genre in genre_list:
            if genre in most_frequent_genres:
                return genre

# apply the function to the cleaned_broad_genre column
streams_df_broader['cleaned_broad_genre'] = streams_df_broader['cleaned_broad_genre'].apply(pick_most_frequent_genre)
```
